```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/tested.csv')
```

```python
df.columns
```

```python
df=df[['Survived','Age','Fare']]
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(df.drop('Survived', axis=1)
,df['Survived'],random_state=0,test_size=0.3)
```

```python
x_train.sample(5)
```

```python
from sklearn.impute import KNNImputer
knn=KNNImputer(n_neighbors=4 ) #n_neighbors default is 5
#weights{‘uniform’, ‘distance’} or callable, default=’uniform’
# Note also reference to KNNImputer
x_train_trf = knn.fit_transform(x_train) #fit  for train
x_test_trf = knn.transform(x_test) # transform for test
```

```python
df_knn = pd.DataFrame(x_train_trf, columns=x_train.columns)
df_knn.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train_trf,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=lr.predict(x_test_trf)
accuracy_score(y_test,y_pred)
```
