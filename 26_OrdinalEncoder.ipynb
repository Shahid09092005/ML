```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Student Attitude and Behavior.csv')
```

```python
df.sample(5)
```

```python
print(df.columns)   # all the columns name by df.columns
```

```python
cols=['prefer to study in','Stress Level ','Financial Status','part-time job']
df=df[cols]
```

```python
df.sample(10)
```

**Train Test Split**

train_test_split function used to split the data into training and testing set

train_test_split focus on that column you want to predict,know as target variable .
in this dataset age and salary are the feature and 'purchase' is a target variable
(use copy for more imformation)

```python
from sklearn.model_selection import train_test_split  #importing the module
x_train , x_test , y_train , y_test = train_test_split(
    df.drop('part-time job' , axis=1),
    df['part-time job'], test_size=0.3,
    random_state=0 )
```

```python
x_train.shape , x_test.shape
```

```python
x_train.nunique()
```

```python
x_train.sample(10)
```

**Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
ord = OrdinalEncoder(categories=[['Morning','Night','Anytime'],['fabulous','Good','Bad','Awful'],['Fabulous','good','Bad','Awful']])
```

```python
#fit the scaler to the train set
ord.fit(x_train)   #learning the values from x_train.

#Transform train and test set
x_train_ord = ord.transform(x_train)
x_test_ord = ord.transform(x_test)
```
