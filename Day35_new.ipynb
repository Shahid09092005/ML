kde for age

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Create a figure for the KDE plots
plt.figure(figsize=(10, 6))

# KDE for 'Age'
sns.kdeplot(df['Age'], color='blue', label='Age', alpha=0.5)

# KDE for 'new_age'
sns.kdeplot(new_df['new_age'], color='orange', label='New Age', alpha=0.5)

# Adding titles and labels
plt.title('KDE Plots of age and new_age')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
